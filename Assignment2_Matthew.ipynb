# Data Model - K nearest neighbours:

__KNN: k = 5, 50% split__

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
for k in range(3,7):
    for j in [0.5, 0.4, 0.2]:
        X_train, X_test, y_train, y_test = train_test_split(VarsModel, Target, test_size = j, random_state = 0)
        clf = KNeighborsClassifier(k, weights = 'distance')
        fit = clf.fit(X_train, y_train)
        y_pre = fit.predict(X_test)
        cm = confusion_matrix(y_test, y_pre)
        print("KNN: k = %d, test = %d percent split") % (k, j*100)
        print cm
        print classification_report(y_test, y_pre, digits=4)
```
